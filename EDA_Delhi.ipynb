```python
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'SO2', 'CO', 'Ozone']

# Group the data by 'site_name' and take the mean of pollutants
# If you want to show the change over time, you could group by both 'site_name' and time
grouped_data = dataset.groupby('site_name')[pollutants].mean()

# Set up the figure for plotting
plt.figure(figsize=(12, 8))

# Plot grouped bar chart using seaborn
grouped_data.plot(kind='bar', width=5, figsize=(14, 8))

# Add chart details
plt.title('Pollutant Concentrations by Site Name', fontsize=16)
plt.ylabel('Concentration (µg/m³)', fontsize=12)
plt.xlabel('Site Name', fontsize=12)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.legend(title='Pollutants', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend location
plt.tight_layout()

# Show the plot
plt.show()
```

```python
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'SO2', 'CO', 'Ozone']

# Group the data by 'site_name' and take the mean of pollutants
grouped_data = dataset.groupby('site_name')[pollutants].mean()

# Set up the figure for plotting
plt.figure(figsize=(16, 8))

# Define bar width (kept the same)
bar_width = 0.8

# Plot grouped bar chart with additional space between bars
grouped_data.plot(kind='bar', width=bar_width, figsize=(16, 8), position=1)

# Adjust x-ticks to space them further apart
plt.xticks(range(len(grouped_data)), grouped_data.index, rotation=45, ha='right')

# Add chart details
plt.title('Pollutant Concentrations by Site Name', fontsize=16)
plt.ylabel('Concentration (µg/m³)', fontsize=12)
plt.xlabel('Site Name', fontsize=12)
plt.legend(title='Pollutants', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend location
plt.tight_layout()

# Show the plot
plt.show()
```
